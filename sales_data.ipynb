```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('shopping_behavior_updated.csv')
df = pd.DataFrame(data).head(30)
df
```

sales relationship bitween season and category

```python

sales_sizes = df.groupby('Category')['Season'].count()
colors=['red','green','blue']
plt.figure(figsize=(10,12))
plt.bar(sales_sizes.index,sales_sizes.values,color=colors)
plt.title('Season sale',size=18,color='red',family='Arial')
plt.xlabel('Category')
plt.ylabel('Sale')
plt.show()
```

```python


sales_sizes = pd.read_csv('shopping_behavior_updated.csv')
sale = pd.DataFrame(sales_sizes).head(30)
sns.pairplot(sale, hue='Season', palette='brg', kind='reg')
```

```python
sns.boxenplot(x='Category',y='Season',data=sale)
```

Top Rated Product


```python
plt.figure(figsize=(14,10))
plt.title('Top Rated Product',color='Green',size=15)
plt.xlabel('Item Purchased',color='blue')
plt.ylabel('Rating',color='blue')
plt.plot(sale['Item Purchased'],sale['Review Rating'],linestyle='--',color='red')
plt.grid()
plt.show()
```

Most used Payment method


```python
plt.figure(figsize=(12,9))
plt.title('Most Used Payment Method',size=15,color='green')
sns.countplot(x='Payment Method',data=sale)
```

Is there any relationship bitween Purchase period and Purchase Amount?

```python
plt.figure(figsize=(12,10))
plt.scatter(sale['Frequency of Purchases'],sale['Purchase Amount (USD)'])
plt.title('Cash spending Mentality',size=15,color='red',family='Arial')
plt.xlabel('Frequency of Purchases',color='green')
plt.ylabel('Purchase Amount (USD)',color='Green')
plt.grid(linestyle='--')
plt.show()
```
